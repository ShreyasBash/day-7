```python
from scipy.stats import binom
import matplotlib.pyplot as plot
```

```python
n = 7
p = 0.2
r_values = list(range(n+1))
dist = [binom.pmf(r,n,p) for r in r_values]
plot.bar(r_values,dist)
plot.show()
```

```python
from scipy.stats import bernoulli
bd = bernoulli(0.7)
x = [0,1]
plot.bar(x,bd.pmf(x))
plot.show()
```

```python
from numpy import random
import seaborn as sb
sb.displot(random.poisson(lam=7,size=10))
plot.show()
```

```python
import numpy as np
exp =np.random.exponential(1,187)
count,bins,ignored = plot.hist(exp,76)
plot.show()
```

```python
mu,sigma = 12,8
s = np.random.normal(mu,sigma,11)
count, bins,ignored = plot.hist(s,231)
plot.plot(bins, 1/sigma*np.sqrt(2*np.pi)*np.exp(-(bins-mu)**2/(2*sigma**2)))
plot.show()
```
